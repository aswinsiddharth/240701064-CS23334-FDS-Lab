```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv('Telco-Customer-Churn.csv')
```

```python
df.head(5)
```

```python
df=df.dropna()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head(5)
```

```python
df.to_csv('Telco-Customer-Churn.csv', index=False)
```

```python
target=df['Churn']
```

```python
df=df.drop('Churn',axis=1)
```

```python
df.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
```

```python
from sklearn import tree
```

```python
model2=tree.DecisionTreeClassifier(random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3=RandomForestClassifier(n_estimators=80)
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
target=target.replace({'Yes':1,'No':0})
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,target,test_size=0.5)
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
c_modelslist = [('logreg', model1), ('dtree', model2),('rforest', model3)]
```

```python
c_model = VotingClassifier(estimators=c_modelslist, voting='soft', n_jobs=-1 )
```

```python
c_model.fit(x_train,y_train)
```

```python
c_model.score(x_test,y_test)
```

```python
c_model.predict(x_test)
```
